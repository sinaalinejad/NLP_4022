## **Input Data**

```python
import pandas as pd
import json
import docx
import pypandoc
from docx import Document as DocxDocument
from hazm import *
import warnings
warnings.filterwarnings('ignore')
from hazm import *

normalizer = Normalizer()

def read_docx(file_path):
    doc = DocxDocument(file_path)
    return "\n".join([para.text for para in doc.paragraphs])

def read_xlsx(file_path):
    df = pd.read_excel(file_path)
    return df

input1 = normalizer.normalize(read_docx('input 1.docx'))
input2 = normalizer.normalize(read_docx('input 2.docx'))

corpus = [input1, input2]

questions1 = read_xlsx('question 1.xlsx')
questions2 = read_xlsx('questions 2.xlsx')

```

## **Preprocessing**

```python
questions1 = questions1.rename(columns = {"سوال":"questions"})
questions1 = questions1['questions']
```

```python
questions2 = questions2[1:]
questions2 = questions2.rename(columns = {"Unnamed: 3":"questions"})
questions2 = questions2['questions']
```

```python
# Convert lists to DataFrame
df1 = pd.DataFrame(questions1, columns=['questions'])
df2 = pd.DataFrame(questions2, columns=['questions'])

appended_series = pd.concat([df1['questions'], df2['questions']], ignore_index=True)

# Convert the series to a DataFrame
result_df = appended_series.to_frame(name='questions')
```

## **LangChain Section**

```python
from langchain.embeddings import SentenceTransformerEmbeddings

embedding_model = SentenceTransformerEmbeddings(model_name='sentence-transformers/LaBSE')
```

```python
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document

# Step 2: Split the document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunk1 = text_splitter.split_text(input1)
chunk2 = text_splitter.split_text(input2)

chunks = chunk1 + chunk2

# Convert the chunks into LangChain Document format
documents = [Document(page_content=chunk) for chunk in chunks]

# Create a Chroma vector store
vector_store = Chroma.from_documents(documents=documents, embedding=embedding_model)


def retrieve_documents(query, top_k=4):
    return vector_store.as_retriever(search_type="similarity", search_kwargs={"k": top_k}).invoke(query)

# Example retrieval
query = 'مواردی که تعداد طبقات، 12 طبقه و بیشتر باشد، منوط به چه چیزی است؟'
retrieved_docs = retrieve_documents(query)
```

## **Text Generation Model + RAG**

### **Load Model**

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, BitsAndBytesConfig
```

```python
def load_model_8b(model_name):
  tokenizer = AutoTokenizer.from_pretrained(model_name)

  model = AutoModelForCausalLM.from_pretrained(
    model_name,
    low_cpu_mem_usage=True,
    load_in_8bit=True,
  )

  return model, tokenizer
```

```python
model, tokenizer = load_model_8b("Qwen/Qwen2-7B-Instruct")
```

```python
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
# model.to(device)
```

### **Final Stage**

```python
from langchain.chains import LLMChain
from langchain.prompts import ChatPromptTemplate

def generate_answer(prompt):
    input_tokens = tokenizer(prompt, return_tensors="pt")
    input_tokens = input_tokens.to(device)
    generate_ids = model.generate(**input_tokens, max_new_tokens=700, do_sample=False, repetition_penalty=1.1)
    model_output = tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]

    return model_output[len(prompt):]

def handle_prompts(context, question):
    prompt_with_context = f"اطلاعات مربوطه: {context}\n\n سوال: {question}\n\n با توجه به اطلاعات داده شده و سوال مطرح شده؛ جواب بده:"
    prompt_without_context = f"سوال: {question}\n\n با توجه سوال مطرح شده؛ جواب بده:"
    answer_with_RAG = generate_answer(prompt_with_context)
    answer_without_RAG = generate_answer(prompt_without_context)

    return answer_with_RAG, answer_without_RAG
```

```python
def rag_system(question):
    retrieved_docs = retrieve_documents(question)
    context = " ".join([doc.page_content for doc in retrieved_docs])
    answer_with_RAG, answer_without_RAG = handle_prompts(context, question)

    return answer_with_RAG, answer_without_RAG

qa_pairs = []
num = 0
# Example usage
for question in result_df['questions']:
    question = question.strip()
    answer_with_RAG, answer_without_RAG = rag_system(question)
    qa_pairs.append({'question': question, 'answer_with_RAG': answer_with_RAG, 'answer_without_RAG': answer_without_RAG})
    print(f"Question: {question}\nAnswer_with_RAG: {answer_with_RAG}\nAnswer_without_RAG: {answer_without_RAG}")
    print('--------------------------\n')
    num+=1

    if num == 20:
      break
```

```python
with open('qa_pairs.json', 'w', encoding='utf-8') as f:
    json.dump(qa_pairs, f, ensure_ascii=False, indent=4)
```
